```python
import numpy as np
import nltk
from nltk.corpus import reuters
import string
import re

def readPairs(fileName):
    f1 = open(fileName, "r")
    pairs = []
    for line in f1:
        pair = line.strip().split('\t')
        for i in len(pair):
            pair[i] = int(pair[i])
        pairs.append(pair)
```

```python
pairs = readPairs('data/pairs.txt')
```
